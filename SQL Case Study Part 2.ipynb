Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT Facilities.name, SUM(IIF(Members.firstname LIKE 'GUEST%' ,Facilities.guestcost*Bookings.slots, Facilities.membercost*Bookings.slots)) AS Revenue
        FROM ((Bookings
        INNER JOIN Members
        ON Bookings.memid=Members.memid)
        INNER JOIN Facilities
        ON Facilities.facid=Bookings.facid)
        GROUP BY Facilities.name
        HAVING SUM(IIF(Members.firstname LIKE 'GUEST%' ,Facilities.guestcost*Bookings.slots, Facilities.membercost*Bookings.slots))<1000
        ORDER BY Revenue;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q12: Find the facilities with their usage by member, but not guests

```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT Facilities.name, Members.surname, Members.firstname, COUNT(Bookings.memid) AS MemberUsage
        FROM ((Bookings
        INNER JOIN Members
        ON Members.memID=Bookings.memid)
        INNER JOIN Facilities
        ON Facilities.facid=Bookings.facid)
        WHERE Bookings.memid != 0
        GROUP BY Facilities.name, Members.surname, Members.firstname
        ORDER BY MemberUsage DESC;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q13: Find the facilities usage by month, but not guests

```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT Facilities.name, strftime('%m',Bookings.starttime) AS Month, COUNT(*) AS MonthlyUsage
        FROM ((Bookings
        INNER JOIN Members
        ON Members.memID=Bookings.memid)
        INNER JOIN Facilities
        ON Facilities.facid=Bookings.facid)
        WHERE Bookings.memid != 0
        GROUP BY Facilities.name, Month;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
